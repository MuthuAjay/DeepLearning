```python
import requests
import os
from tqdm import tqdm

def download_image(row):
    url = row["url"]
    idx = row.name  # Get DataFrame row index

    save_dir = "/home/muthuajay/Documents/GitHub/DeepLearning/datasets/clip/images"
    os.makedirs(save_dir, exist_ok=True)
    
    file_name = os.path.join(save_dir, f"{idx}.jpg")
    
    # Skip if already downloaded
    if os.path.exists(file_name):
        return file_name

    try:
        response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"}, timeout=10)
        response.raise_for_status()  # Raise error for HTTP codes

        with open(file_name, "wb") as f:
            f.write(response.content)

        return file_name

    except Exception as e:
        print(f"Failed to download {url}: {e}")
        return None

```

```python
import requests
import os
from concurrent.futures import ProcessPoolExecutor, as_completed
from tqdm import tqdm

save_dir = "/home/muthuajay/Documents/GitHub/DeepLearning/datasets/clip/images"
os.makedirs(save_dir, exist_ok=True)

def download_image(index_url_tuple):
    idx, url = index_url_tuple
    file_name = os.path.join(save_dir, f"{idx}.jpg")

    if os.path.exists(file_name):
        return idx, file_name  # Already downloaded

    try:
        response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"}, timeout=10)
        response.raise_for_status()
        with open(file_name, "wb") as f:
            f.write(response.content)
        return idx, file_name
    except Exception as e:
        print(f"Failed to download {url}: {e}")
        return idx, None
    
    
# Convert your DataFrame to a list of (index, url) pairs
index_url_list = list(df_10000["url"].items())

results = []

with ProcessPoolExecutor(max_workers=8) as executor:
    futures = [executor.submit(download_image, pair) for pair in index_url_list]
    for future in tqdm(as_completed(futures), total=len(futures)):
        results.append(future.result())

```

```python
# Convert results back to a dict or series
file_path_dict = dict(results)
test_df["file_path"] = test_df.index.map(file_path_dict)
```
